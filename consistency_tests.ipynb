## Persistence analysis
These are big multipliers, and your transition counts are huge (tens of thousands of “current=1” transitions), so this is very unlikely to be noise.

Interpretation: even though episodes are often short, regime membership is much stickier than chance.

```python
import pandas as pd

# Load your existing CSV
df = pd.read_csv("./data/phase2_persistence_stats.csv")

# Inspect columns (optional)
print(df.columns)
print(df)

# Compute ratio explicitly (even if already present)
df["ratio_check"] = df["p_next_given_current"] / df["p_next_baseline"]

# Print results clearly
for _, row in df.iterrows():
    regime = row["regime"]
    p_cond = row["p_next_given_current"]
    p_base = row["p_next_baseline"]
    ratio = row["ratio_check"]

    print(f"{regime.upper()}:")
    print(f"  P(t+1 | t) = {p_cond:.4f}")
    print(f"  Baseline   = {p_base:.4f}")
    print(f"  Ratio      = {ratio:.2f}x\n")
```
